## 数据

```python
from pylib.utils.functions_for_dimentionality_reduction import plot_digits
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
X = digits['data']
y = digits['target']
plot_digits(X, y, seed=1)
```

## T分布随机邻近嵌入

```python
from pylib.utils.functions_for_dimentionality_reduction import plot_digits_projection
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
```

### 投影至二维平面

```python
tsne_2d = TSNE(n_components=2)
X_2d = tsne_2d.fit_transform(X)
```

```python
plot_digits_projection(X_2d, y)
```

### 投影至三维空间

```python
tsne_3d = TSNE(n_components=3)
X_3d = tsne_3d.fit_transform(X)
```

```python
plot_digits_projection(X_3d, y)
```
