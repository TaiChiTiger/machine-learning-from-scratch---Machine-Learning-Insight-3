## 数据

```python
from pylib.utils.functions_for_regression import quadratic_data, plot_scatter
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y, f_truth = quadratic_data(100)
plt.figure(figsize=(8, 6))
plot_scatter(x, y, f=f_truth)
plt.show()
```

## 等张回归

```python
from pylib.utils.functions_for_regression import plot_fitted_line
from sklearn.isotonic import IsotonicRegression
from sklearn.metrics import mean_squared_error
```

```python
ir = IsotonicRegression()
ir.fit(x, y)
```

```python
y_pred = ir.predict(x)
err = mean_squared_error(y, y_pred)
print("样本误差: {:.3f}".format(err)) # 均方差
```

```python
plt.figure(figsize=(10, 8))
plot_scatter(x, y, f=f_truth)
plot_fitted_line(ir, "Isotonic Regression")
plt.show()
```
