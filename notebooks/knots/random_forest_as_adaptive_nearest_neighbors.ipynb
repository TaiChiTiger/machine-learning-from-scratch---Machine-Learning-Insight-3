## 数据

```python
import matplotlib.pyplot as plt
from pylib.utils.functions_for_classification import mixed_2d_2cls, plot_scatter
```

```python
X_train, X_test, y_train, y_test = mixed_2d_2cls()
plt.figure(figsize=(8, 6))
plot_scatter(X_train, y_train)
plt.show()
```

## 随机森林作为自适应最邻近

### 随机森林

```python
from pylib.utils.functions_for_classification import plot_decision_boundary
from sklearn.ensemble import RandomForestClassifier
```

```python
forest = RandomForestClassifier()
forest.fit(X_train, y_train)
```

```python
train_err = 1 - forest.score(X_train, y_train)
print("训练样本误差：{:.3f}".format(train_err))
```

```python
test_err = 1 - forest.score(X_test, y_test)
print("测试样本误差：{:.3f}".format(test_err))
```

```python
# 画出决策边界
plt.figure(figsize=(10, 8))
plot_decision_boundary(X_train, y_train, X_test, y_test, forest)
plt.show()
```

### K邻近

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train, y_train)
```

```python
train_err = 1 - knn.score(X_train, y_train)
print("训练样本误差：{:.3f}".format(train_err))
```

```python
test_err = 1 - knn.score(X_test, y_test)
print("测试样本误差：{:.3f}".format(test_err))
```

```python
# 画出决策边界
plt.figure(figsize=(10, 8))
plot_decision_boundary(X_train, y_train, X_test, y_test, knn)
plt.show()
```
